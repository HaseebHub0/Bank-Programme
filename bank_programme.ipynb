```python

def show_balance(balance):
    print(f"Your balance is PKR{balance:.2f}")
    print()
    print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
    print()

def deposit():
    amount = float(input("Enter the amount to deposit: "))
    print()
    print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
    print()
    if amount < 0:
        print("Invalid amount. Please enter a positive value.")
        return 0
    else:
        return amount


def withdraw(balance):
    amount = float(input("Enter the amount to withdraw: "))
    print()
    print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
    print()
    if amount < 0:
        print("Invalid amount. Amount must be greater than 0.")
        return 0
    elif amount > balance:
        print(f"Insufficient funds. Your balance is PKR{balance:.2f}")
        return 0
    else:
        return amount


def transfer(balance, transaction_history):
    recipient = input("Enter the recipient account number: ")
    amount = float(input("Enter the amount to transfer: "))
    print()
    print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
    print()
    if amount < 0:
        print("Invalid amount. Amount must be greater than 0.")
        return balance
    elif amount > balance:
        print(f"Insufficient funds. Your balance is PKR{balance:.2f}")
        return balance
    else:
        balance -= amount
        transaction_history.append(f"Transferred PKR{amount:.2f} to account {recipient}")
        print(f"You have transferred PKR{amount:.2f} to account {recipient}.")
        print(f"Your new balance is PKR{balance:.2f}")
        print()
        print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
        print()
        return balance


def show_transaction_history(transaction_history):
    print("Transaction History:")
    if transaction_history:
        for transaction in transaction_history:
            print(transaction)
    else:
        print("No transactions found.")
        print()
    print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
    print()

print("Welcome to Bank Programme ğŸ¦")
print("1. Show Balance ğŸ’µ")
print("2. Deposit ğŸ’·")
print("3. Withdraw ğŸ’¸")
print("4. Transfer ğŸ’µ")
print("5. Show Transaction History ğŸ’¹")
print("6. Exit ğŸ™Œ")

def main():
    balance = 0
    transaction_history = []
    is_running = True
    while is_running:

        choice = input("Enter your choice: ")

        if choice == "1":
            show_balance(balance)
        elif choice == "2":
            deposit_amount = deposit()
            if deposit_amount > 0:
                balance += deposit_amount
                transaction_history.append(f"Deposited PKR{deposit_amount:.2f}")
                print(f"You have deposited PKR{deposit_amount:.2f} into your account.")
                print(f"Your new balance is PKR{balance:.2f}")
                print()
                print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
                print()
        elif choice == "3":
            withdraw_amount = withdraw(balance)
            if withdraw_amount > 0:
                balance -= withdraw_amount
                transaction_history.append(f"Withdrew PKR{withdraw_amount:.2f}")
                print(f"You have withdrawn PKR{withdraw_amount:.2f} from your account.")
                print(f"Your new balance is PKR{balance:.2f}")
                print()
                print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
                print()
        elif choice == "4":
            balance = transfer(balance, transaction_history)
            print()
            print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
            print()
        elif choice == "5":
            show_transaction_history(transaction_history)
            print()
            print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
            print()
        elif choice == "6":
            is_running = False
            print("Thank you for using Bank Programme ğŸ¦")
            print("Goodbye!")
            print()
            print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
            print()
        else:
            print("Invalid choice. Please try again.")
            print()
            print("----ğŸ’¸---------------ğŸ’µğŸ’µğŸ’µğŸ’µ-----------------ğŸ’¸----")
            print()


if __name__ == "__main__":
    main()
```
